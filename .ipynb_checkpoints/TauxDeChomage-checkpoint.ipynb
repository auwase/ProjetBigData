```python
# Liste des noms de régions à supprimer
regions_a_supprimer = ['11-Île-de-France', '21-Champagne-Ardenne', '22-Picardie', '23-Haute-Normandie', '24-Centre',
                      '25-Basse-Normandie', '26-Bourgogne', '31-Nord-Pas-de-Calais', '41-Lorraine', '42-Alsace',
                      '43-Franche-Comté', '52-PaysdelaLoire','53-Bretagne', '54-Poitou-Charentes', '72-Aquitaine',
                      '73-Midi-Pyrénées', '74-Limousin', '82-Rhône-Alpes', '83-Auvergne', '91-Languedoc-Roussillon',
                      "93-Provence-Alpes-Côted'Azur",'94-Corse']
```

```python
# Filtrer les lignes en gardant uniquement celles où 'Nom_Depart' n'est pas dans la liste des régions a supprimer 
df_2004_2014 = df_2004_2014[~df_2004_2014['Nom_Depart'].isin(regions_a_supprimer)]
```

```python
df_2004_2014.head(20)
```

```python
# Supprimer les lignes avec des indices 0,1,2,3
df_2004_2014 = df_2004_2014.drop([0,1,2,3], axis=0)
```

```python
# Définir 'Nom_Depart' comme index
df_2004_2014.set_index('Nom_Depart', inplace=True)
```

```python
df_2004_2014.head(20)
```

```python
stats = df_2004_2014.describe()
print(stats)
```

```python
# Sélectionner certains départements pour comparaison
departements_a_comparer = ['75-Paris', '77-Seine-et-Marne', '78-Yvelines', '91-Essonne']

df_2004_2014.loc[departements_a_comparer].T.plot(figsize=(10, 6), marker='o')

plt.title('Évolution du taux de chômage pour certains départements (2004-2014)')
plt.xlabel('Années')
plt.ylabel('Taux de chômage (%)')
plt.legend(title='Départements')
plt.grid(True)
plt.show()
```
