```python
# Vérifier d'abord si les colonnes de df1 et df2 sont identiques

# Étape 1 : Comparer les noms de colonnes
colonnes_identiques = df1.columns.equals(df2.columns)
# Si les colonnes sont identiques, continuer à fusionner, sinon gérer les différences
if colonnes_identiques:
    print("Les colonnes de df1 et df2 sont identiques.")
else:
    print("Attention : les colonnes de df1 et df2 ne sont pas identiques.")
```

```python
# Étape 2 : Comparer les types de colonnes (assurons-nous que chaque colonne a le même type de données)
types_identiques = (df1.dtypes == df2.dtypes).all()
if types_identiques:
    print("Les types de colonnes sont également identiques.")
else:
    print("Attention : certains types de colonnes sont différents entre df1 et df2.")
```

```python
 # Fusionner df1 et df2 (df_combined)
ent_demo= pd.concat([df1, df2], ignore_index=True)
```

```python
# Nettoyage des colonnes pour ent_demo
# Étape 1 : Supprimer les lignes inutiles (0, 1, 2) qui sont inutiles
ent_demo = ent_demo.drop([0, 1, 2])

# Étape 2 : Utiliser la ligne 3 comme noms de colonnes
ent_demo.columns = ent_demo.iloc[0]  # Prendre les noms de colonnes à partir de la ligne 3
ent_demo = ent_demo.drop(3)           # Supprimer la ligne 3 car elle est désormais utilisée pour les noms de colonnes
```

```python

#Créez un nouvel index consécutif pour le DataFrame, en supprimant l'ancien index
# Réinitialiser l'index pour un index consécutif
ent_demo = ent_demo.reset_index(drop=True)

# Nettoyer les noms de colonnes en supprimant les espaces inutiles
ent_demo.columns = ent_demo.columns.str.strip()
```
